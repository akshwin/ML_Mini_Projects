**SPLITTING DATA**

```python
x= df.iloc[:,[5,6,7,9,10]].values
y = df.iloc[:,0].values
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

**BUILDING MODEL**

```python

from sklearn import svm
model = svm.SVC(C=2)
```

```python
model.fit(x_train,y_train)
```

```python
svm_predict = model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_2 = confusion_matrix(y_test, svm_predict)
cm_2
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, classification_report
accuracy= accuracy_score(y_test, svm_predict)
print("Accuracy:", accuracy)

precision, recall, f1score, _ = precision_recall_fscore_support(y_test, svm_predict, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1score)

print("\nClassification Report:")
print(classification_report(y_test, svm_predict))
```
